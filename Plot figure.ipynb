```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
# Apply the default theme
sns.set_theme()
```

```python
data = pd.read_excel('CRT_RF_Result.xlsx', engine='openpyxl')
```

```python
data.columns
```

```python
np.unique(data.dataset.to_numpy())
```

```python
data_ir = data[data['dataset']=='IR']
```

```python
data_ir
```

```python
g = sns.catplot(x="method", y="accuracy",
                hue="split", col="rtt",
                data=data[data['dataset']=='IR'], kind="bar",
                height=4, aspect=.7);
g.set_xlabels('', fontsize=15) # not set_label
g.set_ylabels('Accuracy', fontsize=15)
g.set_xticklabels(size=14)
plt.ylim(90,101)
g.facet_axis(0,0).set_yticklabels(g.facet_axis(0,0).get_yticks(), size=11);
for i in range(5):
    ax = g.facet_axis(0,i)
    for p in ax.patches:
        ax.text(p.get_x() + 0.015, 
                p.get_height() * 1.001, 
                '{0:.2f}'.format(p.get_height()), 
                color='black', rotation='horizontal', size='medium')
plt.savefig('IR.png')
```
